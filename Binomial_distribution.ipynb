```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 10 # number of trails
p= 0.1 # probability of success
size = 1000 # number of samples to generate

binomial_dist = np.random.binomial(n, p, size)

plt.hist(binomial_dist,density=True)
plt.show()
```

I have decreased the probability of success that's why a lot of data move to the left.

### Let's increase the probability of success

```python
n = 10 # number of trails
p = 0.5 # probability of success
size = 1000 # number of samples to generate

binomial_dist = np.random.binomial(n, p, size)

plt.hist(binomial_dist,density=True)
plt.show()
```

This Binomial distribution looks like a normal distribution.

#### Let's increase the probability of success

```python
n = 10 # number of trails
p = 0.9 # probability of success
size = 1000 # number of samples to generate

binomial_dist = np.random.binomial(n, p, size)

plt.hist(binomial_dist,density=True)
plt.show()
```

When we increase the probability of success the Binomial distribution moves to the right.
