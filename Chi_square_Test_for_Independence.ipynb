```python
# WE will use the chi-Square test for independent to see if the survival rate of passengers is independent of the passenger class.
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency
from scipy.stats import chisquare
```

```python
df = pd.read_csv('/content/train.csv')
```

```python
df.head()
```

```python
contingency_table = pd.crosstab(df['Survived'], df['Pclass'])
contingency_table
```

```python
# Perform the Chi-Square test
chi2, p_value, dof, expected = chi2_contingency(contingency_table)

# Print the results
print('\nChi-square statistic: {:.2f}', format(chi2))
print('P-value statistic: {:.2f}', format(p_value))
print('Degree of freedom statistic: {:.2f}', format(dof))
print('Expected Frequencies statistic: \n{:.2f}', format(expected))
```
