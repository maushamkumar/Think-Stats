```python
import pandas as pd

df = pd.read_csv('/content/train.csv')
df.head()
```

```python
# Count passengers in each class
class_counts = df['Pclass'].value_counts().sort_index()
class_counts
```

The above code give our observed data and our expected data is 297

```python
# calcaulate the expected counts assuming a uniform distribution
total_passangers = len(df)
expected_counts = total_passangers/3
expected = [expected_counts]*3
print("Expected passenger counts (assuming uniform distribution): \n", expected)
```

```python
# Perform the chi-square Goodness of fit test
import scipy.stats as stats
chi2, p_value = stats.chisquare(class_counts, expected)

# print the results
print("\nChi-square Statistic: {:.2f}".format(chi2))
print("p_value: {:.4f}".format(p_value))
```

```python
# Interpret the results
alpha = 0.05
if p_value < alpha:
  print("\nWe reject the null hypothesis. The distribution of passengers among the classes is not uniform.")
else:
  print("\nWe fail to reject the null hypothesis. The distribution of passengers among the classes is uniform. ")
```
